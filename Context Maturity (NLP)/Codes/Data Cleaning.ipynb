# Loading Data and Preparing Texts for Cleaning
--- 


```python
#Importinig modules
import os
import pandas as pd 
import numpy as np
import re
import string

#Loading data
directory_path = '/Users/qab/Desktop/Personal/NLP Projects/Context Maturity (NLP)/Data'
music_files = os.listdir(directory_path + '/translation through speakers')

tts = []
#tts is the name for the album; translation through speakers
for i, music in enumerate(music_files):
    with open(directory_path + '/translation through speakers/' + music, 'rt') as file:
        tts.append(file.read())

songs_dictionary = dict(zip(music_files, tts))

#Combining list of texts and putting it into a pandas dataframe
def combine_text(list_of_text):
    #Takes a list of text and combines them into one large chunk of text.
    combined_text = ''.join(list_of_text)
    return combined_text

data_combined = {key: [combine_text(value)] for (key, value) in songs_dictionary.items()}

pd.set_option('max_colwidth', -1)

tts = pd.DataFrame.from_dict(data_combined).transpose()
tts.columns = ['lyrics']
tts = tts.sort_index()
tts = pd.DataFrame.reset_index(tts, drop = False)
tts.rename(columns = {'index':'titles'}, inplace = True)
```

## Notes about lyrics
---
* They are plain texts
* They can have repeated lines which can affect the effectiveness of our analysis
* Music can contain metaphors and other literary devices which can affect the semantics of the texts
* There are no obvious typos or mistakes
* There are punctuations, square brackets, artist names, verse numbers and some errors in the text like (\n)
* The mistake above should be replaced with space(' ')
* There are dashes between some words
* 

```python
titles = ['dead man wallking', 'dont ask cuz i dont know', 'for the dreamers', 'life', 'paper planes', 'the wonder years', 'timeless', 'waves of loneliness', 'while you count sheep', 'wutup snow']
```

```python
#Defining a function for initial text cleaning

def clean(lyrics):
#The initial cleaning involves removing punctuation, texts in square brackets and making lyrics lowercases
    lyrics = re.sub('\[.*?\]', '', lyrics)
    lyrics = re.sub('[%s]' % re.escape(string.punctuation), '', lyrics)
    lyrics = re.sub('\n', ' ', lyrics)
    lyrics = re.sub('[''""...]', '', lyrics)
    lyrics = lyrics.lower()
    return lyrics
cleaning = lambda song: clean(song)
```

```python
translation_through_speakers = pd.DataFrame(tts.lyrics.apply(cleaning))
translation_through_speakers
```

```python
translation_through_speakers['titles'] = titles
translation_through_speakers
```
